**DATASET DEBUNK**

BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)

PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)

default.payment.next.month: Default payment (1=yes, 0=no)

```python
df["pay_mean"] = df[["PAY_0","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"]].mean(axis=1)
df["pay_max"]  = df[["PAY_0","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"]].max(axis=1)
df["pay_std"]  = df[["PAY_0","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"]].std(axis=1)
```

```python
df=df.drop(columns=["ID"])
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_auc_score, f1_score
```

```python
X=df.drop(columns=["default.payment.next.month"])
y=df['default.payment.next.month']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8, stratify=y)
```

```python
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

estimators = [
    ('clf', XGBClassifier(random_state=8))
]
pipe = Pipeline(steps=estimators)
pipe
```

```python
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

search_space = {
    'clf__n_estimators' : Integer(2,20),
    'clf__max_depth': Integer(2,8),
    'clf__learning_rate': Real(0.001, 1.0, prior='log-uniform'),
    'clf__subsample': Real(0.5, 1.0),
    'clf__colsample_bytree': Real(0.5, 1.0),
    'clf__colsample_bylevel': Real(0.5, 1.0),
    'clf__colsample_bynode' : Real(0.5, 1.0),
    'clf__reg_alpha': Real(0.0, 10.0),
    'clf__reg_lambda': Real(0.0, 10.0),
    'clf__gamma': Real(0.0, 10.0)
}

opt = BayesSearchCV(pipe, search_space, cv=3, n_iter=10, scoring='roc_auc', random_state=8)
```

```python
opt.fit(X_train, y_train)
```

```python
opt.best_score_
```

```python
opt.score(X_test,y_test)
```

```python
y_pred = opt.predict(X_test)
y_proba = opt.predict_proba(X_test)[:, 1]

print(" Confusion Matrix: \n", confusion_matrix(y_test, y_pred))

auc = roc_auc_score(y_test, y_proba)
print(f"\n AUC: {auc:.4f} \n")

f1 = f1_score(y_test, y_pred)
print(f" F1-score: {f1:.4f} ")
```
